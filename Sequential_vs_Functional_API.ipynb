# Keras Sequential and Functional APIs for Building Neural Networks

**Creating a neural network using Sequential API**

```python
!pip install ann_visualizer
```

```python
# importing necessary libraries
from tensorflow import keras
# for visualizing created Artificial Neural Network (ANN)
from ann_visualizer.visualize import ann_viz
```

```python
# creating a neural network using keras Sequential API
seq_model = keras.Sequential()
seq_model.add(keras.layers.Dense(8, input_shape = (8,), activation = "relu"))
seq_model.add(keras.layers.Dense(4, activation = "relu"))
seq_model.add(keras.layers.Dense(1, activation = "sigmoid"))
# getting the model summary
seq_model.summary()
```

```python
# visualizing the ANN
ann_viz(seq_model, view = True,  title="Neural Network Created using Sequential API")
```

**Creating a neural network using Functional API**

```python
# importing libraries
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Model
```

```python
# creating the model - Functional
input_layer = Input(shape = (8,))
layer1 = Dense(8, input_shape = (8,), activation = "relu")(input_layer)
layer2 = Dense(4, activation = "relu")(layer1)
output_layer = Dense(1, activation = "sigmoid")(layer2)
func_model = Model(inputs = [input_layer], outputs = [output_layer])
# getting model summary
func_model.summary()
```

**Note that both model parameters are same**
